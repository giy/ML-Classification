## Filter the products

```python
filtered_products = products[products['rating'] != 3]
```

# Add sentiment column

```python
filtered_products['sentiment'] = filtered_products['rating']>= 4
```

```python
graphlab.canvas.set_target('ipynb')
```

```python
filtered_products.head()
```

# Train a sentiment model containing the selected words. Just like we'd chosen zip_code, sqft_living etc for zip_code, here we select our filtered_words

```python
train_data, test_data = filtered_products.random_split(.8, seed=0)
```

```python
selected_words_model = graphlab.logistic_classifier.create(train_data, features=selected_words, 
                                                           target='sentiment', validation_set=test_data)
```

```python
selected_words_model['coefficients'].print_rows(num_rows=12)
```

```python
selected_words_model.evaluate(test_data)
```

```python
selected_words_model.evaluate(test_data, metric='roc_curve')
```

```python
selected_words_model.show(view='Evaluation')
```
